```python
# original simple server to only receive data
# has a delay after receiving after certain amount of data

import socket
import base64

s = socket.socket()
s.bind(('192.168.29.193', 5000))
s.listen(0)

try:
    while True:
        client, addr = s.accept()

        while True:
            content = client.recv(32)
            if len(content) == 0:
                break
            else:
                # Decode Base64 and print the result
                decoded_data = base64.b64decode(content)
                mean = decoded_data.decode('utf-8')
                print(mean)
        
        client.close()
except KeyboardInterrupt:
    print("Closing connection")
```

```python
# Receive data on server and save 10 sec of wav file

import socket
import base64
import wave
import numpy as np

s = socket.socket()
s.bind(('192.168.29.193', 5000))
s.listen(0)

# Set the sample width, sample rate, and channels for the WAV file
sample_width = 2  # 16-bit audio
sample_rate = 44100  # 44.1 kHz
channels = 1  # Mono

fileSwitch = False

# Calculate the number of frames needed for 10 seconds
duration = 10  # seconds
num_frames = int(sample_rate * duration)

try:
    while True:
        client, addr = s.accept()

        # Initialize a NumPy array to store the audio data
        audio_data = np.array([], dtype=np.int16)

        # Calculate the number of frames needed for 10 seconds
        num_frames = int(sample_rate * duration)

        # Buffer size is adjusted to be a multiple of sample width (16 bits = 2 bytes)
        buffer_size = 64

        while len(audio_data) < num_frames:
            content = client.recv(buffer_size)

            if len(content) == 0:
                break
            else:
                # Decode Base64 and append the data to the NumPy array
                decoded_data = base64.b64decode(content)
                audio_chunk = np.frombuffer(decoded_data, dtype=np.int16)
                audio_data = np.concatenate((audio_data, audio_chunk))

        # Trim or zero-pad the audio data to ensure it has the required number of frames
        audio_data = audio_data[:num_frames]
        if len(audio_data) < num_frames:
            audio_data = np.concatenate((audio_data, np.zeros(num_frames - len(audio_data), dtype=np.int16)))

        # Save the audio data to a WAV file
        if not fileSwitch:
            output_file = 'output1.wav'
            fileSwitch = True
        else:
            output_file = 'output2.wav'
            fileSwitch = False

        with wave.open(output_file, 'w') as wave_file:
            wave_file.setnchannels(channels)
            wave_file.setsampwidth(sample_width)
            wave_file.setframerate(sample_rate)
            wave_file.writeframes(audio_data.tobytes())

        print(f"Audio data saved to {output_file}")

        client.close()
except KeyboardInterrupt:
    print("Closing connection")
```
